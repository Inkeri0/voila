```python
def select_country_data(Country, health_factors):

  return health_factors.loc[health_factors['Country'] == Country, ]
```

```python
health_factors_full_data.head()
```

Creating the charts

```python
def visualize_country_data(Country, health_factors_full_data):
    data_to_viz = health_factors_full_data[health_factors_full_data['Country'] == Country]

    # These are the ranged we need to make a better looking graph, with these ranges the data is way clearer to visuale
    gdp_range = (data_to_viz['GDP'].min(), data_to_viz['GDP'].max())
    life_expectancy_range = (data_to_viz['Life expectancy'].min(), data_to_viz['Life expectancy'].max())
    income_range = (data_to_viz['Income composition of resources'].min(), data_to_viz['Income composition of resources'].max())

    chart = alt.Chart(data_to_viz).mark_point().encode(
        x=alt.X('Life expectancy:Q', scale=alt.Scale(domain=life_expectancy_range), title='Life Expectancy'),
        y=alt.Y('Income composition of resources:Q', scale=alt.Scale(domain=income_range), title='Income Composition of Resources'),
        color=alt.Color('GDP:Q', scale=alt.Scale(domain=gdp_range), title='GDP')
    ).properties(
        title=f'{Country}: Relationships between GDP, Life Expectancy, and Income Composition of Resources'
    )

    display(chart)
```

A graph visualising the relationship between GDP, Life expectancy and income Composition of Resources from a chosen country in the code.

```python
visualize_country_data('Afghanistan', health_factors_full_data)
```

```python
dropdown = widgets.Dropdown(
    options=(health_factors['Country'].unique()),
    description='Country:',
)
```

Interactive, user can choose a country from the dropdown menu and the Relationhip between GDP, Life Expectancy and Income Composition of Resources will show up respectivly

```python
from ipywidgets import Output


output = Output()

def on_dropdown_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        with output:
            clear_output(wait=True)
            visualize_country_data(change['new'], health_factors_full_data)

dropdown.observe(on_dropdown_change)

display(dropdown)
display(output) #Had to add this otherwise not visible in binder voila
```

Conclusion: Based on our charts, we can see that as GDP increases, so does life expectancy. Similarly, when income composition improves, life expectancy also rises.
